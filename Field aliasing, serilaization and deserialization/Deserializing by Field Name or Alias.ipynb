```python
from pydantic import BaseModel, ConfigDict, Field, ValidationError
from typing import Optional
```

```python
class Model(BaseModel):
    id_: int = Field(alias="id")
    first_name: str = Field(alias="firstName")
```

```python
try:
    Model(id_=10, first_name="Newton")
except ValidationError as ex:
    print(ex)
```

```python
data = {
    "id_": 10,
    "first_name": "Newton"
}

try:
    Model.model_validate(data)
except ValidationError as ex:
    print(ex)    
```

```python
class Model(BaseModel):
    model_config = ConfigDict(populate_by_name=True)
    
    id_: int = Field(alias="id")
    first_name: str = Field(alias="firstName")
```

```python
Model(id_=10, first_name="Newton")
```

```python
Model(id_=10, firstName="Newton")
```

```python
data = {
    "id_": 10,
    "first_name": "Newton"
}

Model.model_validate(data)
```

```python
from pydantic.alias_generators import to_camel

class Person(BaseModel):
    model_config = ConfigDict(
        alias_generator=to_camel,
        populate_by_name=True,
        extra="forbid"
    )

    id_: int = Field(alias="id", default=1)
    first_name: Optional[str] = None
    last_name: str
    age: Optional[int] = None
```

```python
p = Person(id=10, first_name='Isaac', lastName='Newton', age=84)
p
```

```python
data_json = """
{
    "id": 10,
    "firstName": "Isaac",
    "last_name": "Newton",
    "age": 84
}
"""

p = Person.model_validate_json(data_json)
p
```

```python
p.model_dump()
```

```python
p.model_dump(by_alias=True)
```

```python
p.model_dump_json(by_alias=True)
```
