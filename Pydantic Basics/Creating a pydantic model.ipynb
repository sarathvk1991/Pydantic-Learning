```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    first_name:str
    last_name:str
    age:int
```

```python
p = Person(first_name="Sarath", last_name="V", age=34)
```

```python
class Person(BaseModel):
    first_name:str
    last_name:str
    age:int

    @property
    def display_name(self):
        return f"{self.first_name} {self.last_name}"
```

```python
p = Person(first_name="Sarath", last_name="V", age=34)
p.display_name
```

```python
try:
    p1 = Person(first_name="Sarath", last_name="V", age="thirtyfour")
except ValidationError as e:
    print(e)
```
