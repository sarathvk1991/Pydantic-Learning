```python
from pydantic import BaseModel, ValidationError

class Person(BaseModel):
    first_name:str
    last_name:str
    age:int
```

```python
p1 = Person(first_name= "Person", last_name="A", age = 10)
p2 = Person(first_name= "Person", last_name="B", age = 20)
```

```python
p1.model_dump() #returns dict
```

```python
p2.model_dump_json() #returns json string
```

```python
p2.model_dump_json(exclude=['age'])
```

```python
p2.model_dump(include=['age'])
```
