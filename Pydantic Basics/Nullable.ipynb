```python
from pydantic import BaseModel, ValidationError
from typing import Optional,Union
class Model(BaseModel):
    a:Optional[int] #or you can use Union by importing from typing, then it would be a:Union[int,None]
    #a: int | None # This only works in python 3.10+
```

```python
Model(a=None)
```

```python
Model(a=10)
```

```python
try:
    Model()
except ValidationError as e:
    print(e)
```

```python
class Model(BaseModel):
    a:Union[int, None] = None
```

```python
Model()
```

```python
#We can make the field 
# 1. Optional and Nullable by giving a default value and using Union, pipe or optional
# 2. Optional and not nullable by giving a default value
# 3. Required and Nullable by using Union, pipe  or optio
```
