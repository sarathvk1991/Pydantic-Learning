```python
from pydantic import BaseModel, ValidationError
```

```python
class Circle(BaseModel):
    center: tuple[int,int] = (0,0) #Which will make this field optional with a default value. Even giving ("junk") as default will work here
    radius: int
```

```python
Circle.model_fields
```

```python
Circle(radius=2)
```

```python
data = {"radius":1}
Circle.model_validate(data)
```

```python
data_json = '''
{
"radius":"1"
}
'''
Circle.model_validate_json(data_json)
```

```python
Circle(center=(1,1),radius=29)
```
