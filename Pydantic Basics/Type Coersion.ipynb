```python
from pydantic import BaseModel, ValidationError
```

```python
class Coordinates(BaseModel):
    x:float
    y:float
```

```python
c1 = Coordinates(x=2.5, y=5.5)
c1
```

```python
c2 = Coordinates(x=0,y="3.5")
c2
```

```python
class Model(BaseModel):
    x:str
```

```python
Model(x="demo")
```

```python
try:
    Model(x=100)
except ValidationError as e:
    print(e)
```

```python
new_data = '''{
"x": {
"personal":"sarath@personal.com",
"work":"sarath@work.com"
}
}
'''
```

```python
try:
    Model.model_validate_json(new_data)
except ValidationError as e:
    print(e)
```

```python
new_data_dict = {
"x": {
"personal":"sarath@personal.com",
"work":"sarath@work.com"
}
}
```

```python
Model(x=str(new_data_dict['x']))
```
