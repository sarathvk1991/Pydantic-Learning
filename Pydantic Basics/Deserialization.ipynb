```python
from pydantic import BaseModel, ValidationError
```

```python
class Person(BaseModel):
    first_name:str
    last_name:str
    age:int
```

```python
p = Person(first_name="Sarath", last_name="V", age=35) #Direct
data = {"first_name":"Sarath", "last_name":"V", "age":"35"}
p1 = Person.model_validate(data) #Using Dictionary
data1 = {"first_name":"Sarath", "last_name":"V"}
try:
    p2 = Person.model_validate(data1) #Missing Data Error
except ValidationError as e:
    print(e)
json_data = '''
{"first_name":"Sarath", "last_name":"V", "age":"35"}
'''
p3 = Person.model_validate_json(json_data) # Using Json String
```
