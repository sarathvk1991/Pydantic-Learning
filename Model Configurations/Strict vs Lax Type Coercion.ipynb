```python
from pydantic import BaseModel, ValidationError,ConfigDict
from typing import Optional,Set,List
```

```python
class Model(BaseModel):
    field_1:str
    field_2:float
    field_3:list
    field_4:tuple
```

```python
try:
    Model(field_1=10, field_2=1, field_3=(1,2,3), field_4=[1,2,3])
except ValidationError as e:
    print(e)
```

```python
Model(field_1="10", field_2=1, field_3=(1,2,3), field_4=[1,2,3,4])
```

```python
class Model(BaseModel):
    model_config = ConfigDict(strict=True)
    
    field_1: bool
    field_2: float
    field_3: int
    field_4: Optional[str]
    field_5: tuple[int, ...]
    field_6: Set[str]
    field_7: dict
```

```python
data = Model.model_validate_json(json_data)
```

```python
json_data = '''
{
    "field_1": true,
    "field_2": 10,
    "field_3": 1,
    "field_4": null,
    "field_5": [1, 2, 3.5],
    "field_6": ["a", "b", 100],
    "field_7": {"a": 1, "b": 2}
}
'''

try:
    Model.model_validate_json(json_data)
except ValidationError as ex:
    print(ex)
```
