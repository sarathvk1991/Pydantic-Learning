```python
from pydantic import BaseModel, ConfigDict, ValidationError

class Model(BaseModel):
    field_1: int = None
    field_2: str = 100
```

```python
try:
    Model(field_1=None, field_2=100)
except ValidationError as ex:
    print(ex)
```

```python
class Model(BaseModel):
    model_config = ConfigDict(validate_default=True)
    
    field_1: int = None
    field_2: str = 100
```

```python
try:
    Model()
except ValidationError as ex:
    print(ex)
```

```python
class Model(BaseModel):
    field: int
```

```python
m = Model(field=10)
m.field = "Python"
m
```

```python
class Model(BaseModel):
    model_config = ConfigDict(validate_assignment=True)
    
    field: int
```

```python
m = Model(field=10)
```

```python
try:
    m.field = "Python"
except ValidationError as ex:
    print(ex)
```
