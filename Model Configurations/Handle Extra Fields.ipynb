```python
from pydantic import BaseModel, ValidationError, ConfigDict
```

```python
class Model(BaseModel):
    model_config = ConfigDict(extra="forbid")#raises an error, ignore is default which just ignores
    field_1: int
```

```python
try:
    m= Model(field_1=10,field_2=20)
except ValidationError as e:
    print(e)
```

```python
class Model(BaseModel):
    model_config = ConfigDict(extra="allow")#allows any additional field
    field_1: int
```

```python
m= Model(field_1=10,field_2=20)
m
```

```python
m.model_extra #shows field_2
```
