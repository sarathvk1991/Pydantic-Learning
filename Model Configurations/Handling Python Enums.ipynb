```python
from enum import Enum
```

```python
class Color(Enum):
    red = "Red"
    green = "Green"
    blue = "Blue"
    orange = "Orange"
    yellow = "Yellow"
    cyan = "Cyan"
    white = "White"
    black = "Black"
```

```python
Color.orange.value
```

```python
from pydantic import BaseModel, ConfigDict, ValidationError
```

```python
class Model(BaseModel):
    color: Color
```

```python
Model(color=Color.red)
```

```python
data = """
{
    "color": "Red"
}
"""

m=Model.model_validate_json(data)
m.model_dump_json()
```

```python
data = """
{
    "color": "Magenta"
}
"""

try:
    Model.model_validate_json(data)
except ValidationError as ex:
    print(ex)
```

```python
class Model(BaseModel):
    model_config = ConfigDict(use_enum_values=True)

    color: Color
```

```python
m = Model(color=Color.cyan)
m
```

```python
data = """
{
    "color": "Magenta"
}
"""

try:
    Model.model_validate_json(data)
except ValidationError as ex:
    print(ex)
```
