```python
from pydantic import (
    BaseModel, 
    ConfigDict, 
    PastDatetime, 
    PastDate, 
    AwareDatetime, 
    NaiveDatetime, 
    ValidationError,
)
```

```python
from datetime import datetime, timedelta

import pytz

local_one_hour_ago = datetime.now() - timedelta(hours=1)
```

```python
utc_one_hour_ago = local_one_hour_ago.astimezone(pytz.utc)
```

```python
utc_naive_one_hour_ago = utc_one_hour_ago.replace(tzinfo=None)
```

```python
class Model(BaseModel):
    dt: PastDatetime
```

```python
m = Model(dt=local_one_hour_ago)
m
```

```python
m = Model(dt=utc_one_hour_ago)
m
```

```python
m = Model(dt=utc_naive_one_hour_ago)
m
```

```python
class Model(BaseModel):
    dt: NaiveDatetime
```

```python
Model(dt=local_one_hour_ago)
```

```python
try:
    Model(dt=utc_one_hour_ago)
except ValidationError as ex:
    print(ex)
```

```python
class Model(BaseModel):
    dt: AwareDatetime
```

```python
Model(dt=utc_one_hour_ago)
```

```python
try:
    Model(dt=local_one_hour_ago)
except ValidationError as ex:
    print(ex)
```
