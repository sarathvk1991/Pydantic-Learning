```python
from uuid import uuid4
```

```python
from pydantic import BaseModel, Field, UUID4, ValidationError
```

```python
class Person(BaseModel):
    id: UUID4
```

```python
p = Person(id=uuid4())
p
```

```python
class Person(BaseModel):
    id: UUID4 = uuid4()
```

```python
p1 = Person()
```

```python
p2 = Person()
```

```python
p1.model_dump()
```

```python
p2.model_dump()
```

```python
class Person(BaseModel):
    id_: UUID4 = Field(alias="id", default=uuid4())
```

```python
p1 = Person()
```

```python
p2 = Person()
```

```python
p1.model_dump()
```

```python
p2.model_dump()
```

```python
class Person(BaseModel):
    id_: UUID4 = Field(alias="id", default_factory=uuid4)
```

```python
p1 = Person()
```

```python
p2 = Person()
```
