```python
from pydantic import BaseModel, Field, ValidationError
```

```python
from math import pi


class Circle(BaseModel):
    center: tuple[int, int] = (0, 0)
    radius: int = Field(default=1, gt=0)

    def area(self):
        return pi * self.radius ** 2
```

```python
c = Circle(center=(1, 1), radius=2)
```

```python
c.model_dump()
```

```python
c.area()
```

```python
class Circle(BaseModel):
    center: tuple[int, int] = (0, 0)
    radius: int = Field(default=1, gt=0)

    @property
    def area(self):
        return pi * self.radius ** 2
```
