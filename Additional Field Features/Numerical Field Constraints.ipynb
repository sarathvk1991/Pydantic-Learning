```python
from pydantic import BaseModel, Field, ValidationError
```

```python
class Model(BaseModel):
    number: float = Field(gt=2, le=10, multiple_of=2)
```

```python
Model(number=10)
```

```python
try:
    Model(number=3)
except ValidationError as ex:
    print(ex)
```

```python
try:
    Model(number=14)
except ValidationError as ex:
    print(ex)
```

```python
class Model(BaseModel):
    number: int = Field(gt=0)
```

```python
Model.model_fields
```

```python
from pydantic import PositiveInt

class Model(BaseModel):
    number: PositiveInt
```

```python
Model.model_fields
```
