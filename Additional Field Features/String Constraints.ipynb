```python
from pydantic import BaseModel, Field, ValidationError
```

```python
class Model(BaseModel):
    name: str = Field(min_length=1, max_length=20)
```

```python
Model(name="Richard")
```

```python
try:
    Model(name="")
except ValidationError as ex:
    print(ex)
```

```python
try:
    Model(name="*" * 21)
except ValidationError as ex:
    print(ex)
```

```python
class Model(BaseModel):
    items: list[float] = Field(min_length=3, max_length=5, default=[1.0, 2.0, 3.0])
```

```python
Model()
```

```python
Model(items=[2, 3.5, 4])
```

```python
try:
    Model(items=[1, 2])
except ValidationError as ex:
    print(ex)
```

```python
class Circle(BaseModel):
    center: tuple[int]
    radius: int = Field(gt=0, default=1)
```

```python
try:
    Circle(center=(1, 1), radius=1)
except ValidationError as ex:
    print(ex)
```

```python
class Circle(BaseModel):
    center: tuple[int, ...]
    radius: int = Field(gt=0, default=1)
```

```python
Circle(center=(1, 1))
```

```python
Circle(center=(1, 1, 1))
```

```python
Circle(center=(1, 2, 3, 4, 5))
```

```python
class Circle(BaseModel):
    center: tuple[int, ...] = Field(min_length=2, max_length=3, default=(0, 0))
    radius: int = Field(gt=0, default=1)
```
